```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('creditcard.csv')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df['Class'].unique()
```

```python
df['Class'].value_counts()
```

```python
df.groupby('Class').mean()
```

```python
legal=df[df['Class']==0]
fraud=df[df['Class']==1]
```

```python
print(legal.shape,fraud.shape)
```

```python
legal_samp=legal.sample(n=492) #no. of legal samples shud be same as fraudulant samples
legal_samp.shape
```

```python
legal_fraud=pd.concat([legal_samp,fraud],axis=0)
legal_fraud.shape
legal_fraud.head(5)
```

```python
legal_fraud['Class'].value_counts()
```

```python
# Scaling the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(legal_fraud.drop(columns='Class'))
scaled_legal_fraud_df = pd.DataFrame(scaled_data, columns=legal_fraud.columns[:-1])
```

```python
scaled_legal_fraud_df.sample(5)
```

```python
X = scaled_legal_fraud_df
y = legal_fraud['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```

```python
print("Training Data Accuracy: ",accuracy_score(train_predict, y_train))
```

```python
print("testing Data Accuracy: ",accuracy_score(test_predict, y_test))
```
